# Câu hỏi 1: Tính độ dài của vector (D)

```python
import numpy as np

def compute_vector_length(vector):
    return np.sqrt(np.sum(vector**2))

vector = np.array([-2, 4, 9, 21])
result = compute_vector_length(vector)
print(round(result, 2))
```

# Câu hỏi 5

```python
import numpy as np

def compute_cosine_similarity(v1, v2):
    return np.dot(v1, v2)/((np.linalg.norm(v1))*(np.linalg.norm(v2)))

x = np.array([1, 2, 3, 4])
y = np.array([1, 0, 3, 0])
result = compute_cosine_similarity(x, y)
print(round(result, 3))
```

# **Background Subtraction**

```python
import cv2
import matplotlib.pyplot as plt

bg1_image = cv2.imread('GreenBackground.png', 1)
ob_image = cv2.imread('Object.png', 1)
bg2_image = cv2.imread('NewBackground.jpg', 1)
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(cv2.cvtColor(bg1_image, cv2.COLOR_BGR2RGB))
axs[0].set_title("Green Background")
axs[0].axis("off")

axs[1].imshow(cv2.cvtColor(ob_image, cv2.COLOR_BGR2RGB))
axs[1].set_title("Object")
axs[1].axis("off")

axs[2].imshow(cv2.cvtColor(bg2_image, cv2.COLOR_BGR2RGB))
axs[2].set_title("New Background")
axs[2].axis("off")

plt.tight_layout()
plt.show()
```

```python
#RESIZE
IMG_SIZE = (678, 381)
bg1_image = cv2.resize(bg1_image, IMG_SIZE)
ob_image = cv2.resize(ob_image, IMG_SIZE)
bg2_image = cv2.resize(bg2_image, IMG_SIZE)
print(bg1_image.shape)
print(ob_image.shape)
print(bg2_image.shape)
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

axs[0].imshow(cv2.cvtColor(bg1_image, cv2.COLOR_BGR2RGB))
axs[0].set_title("Green Background")
axs[0].axis("off")

axs[1].imshow(cv2.cvtColor(ob_image, cv2.COLOR_BGR2RGB))
axs[1].set_title("Object")
axs[1].axis("off")

axs[2].imshow(cv2.cvtColor(bg2_image, cv2.COLOR_BGR2RGB))
axs[2].set_title("New Background")
axs[2].axis("off")

plt.tight_layout()
plt.show()
```

```python
#LẤY MASK CỦA OBJECT
def compute_difference(bg_img, input_img):
    return np.abs(bg_img - input_img).mean(axis=2) #tính trung bình trên channel thứ 3

difference_single_channel = compute_difference(bg1_image, ob_image)
difference_single_channel
print(difference_single_channel)
plt.imshow(difference_single_channel, cmap='gray')
plt.title("Difference")
plt.axis("off")
plt.show()
```

```python
def compute_binary_mask(difference_single_channel):
    difference_binary = np.where(difference_single_channel >= 15, 255, 0)
    difference_binary = np.stack((difference_binary,)*3, axis=-1)
    return difference_binary

binary_mask = compute_binary_mask(
    difference_single_channel
)
plt.imshow(binary_mask)
plt.title("Binary")
plt.axis("off")
plt.show()
```

```python
def replace_background(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(bg1_image,ob_image)
    binary_mask = compute_binary_mask(difference_single_channel)

    output = np.where(binary_mask==255, ob_image, bg2_image)

    return output

output = replace_background(bg1_image, bg2_image, ob_image)

plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.title("Binary")
plt.axis("off")
plt.show()
```
